#Exploring Variables Relationship

Now that we understood how to make histograms or bar plots in order to show physically what are the variables we're studying one after the other, we will study how to show 2 different variables within the same illustration.



# Why should we analyse 2 variables together ?

We want to analyse 2 different variables in order to **put them into perspective** ! For example, we can try to compare 2 different continuous variables, such as **age** and a **feeling thermometer** for an election candidate. 

# How to compare 2 variables ? 

Now let's see how to make it, it can be very useful for researches purposals, to make articles using the data we can find in ANES in order to analyse the 2020 American Elections. 

First: **Importing pandas** in our document is mandatory

```
import pandas as pd
my_link = "https://raw.githubusercontent.com/datamisc/ts-2020/main/data.csv"
```

As the file is too heavy, it's necessary to **compress** it

```
df=pd.read_csv(my_link, compression='gzip')
```

Now, as we import pandas and the database which we are interested in, we can assign 2 different variables in the same representation. To analyse the relationship between vote and age, we can make a double research, using 3 different variables to analyse if there is a correlation or not, concerning the 2020 American Elections. Therefore, we can use 2 different feeling thermometers (one for Biden, one for Trump), and the variable "Respondent's age". WARNING: about the interpretation of the results --> It's a **feeling thermometer**, and not a variable making us aware of which respondent voted for which candidate.

```
biden_variables = ["V202143", "V201507x"]
biden_df = df[biden_variables]
biden_df.columns = ["biden", "age"]
```

Now, we can make a **scatter plot** with these 2 variables to check if we can understand something about the relationship between them !

```
biden_df.plot(kind= "scatter", x= "age", y= "biden")
```

Let's see this scatter plot. We've got several difficulties to understand it. First, we can change something about the dots' transparency, using the **alpha parameter**. 

```
biden_df.sample(500).plot(kind="scatter", x="age", y="biden", alpha= 0.3)
```

Making the dots more transparents allows us to understand where is the majority of them. However, something is still problematic. Indeed, we've got several **dots below the 0 value**. Let's delete them.

```
biden_filter= biden_df["biden"]>=0
print(biden_filter)
biden_df= biden_df.loc[biden_filter]
print(biden_df)
```

Now, we deleted them, let's check it out.

```
biden_df.plot(kind="scatter", x="age", y="biden", alpha=0.3)
```

About the Biden's feeling thermometer, that's better now. But there is still, on the right side of the scatter, a problem concerning the age. Let's fix it.

```
biden_newfilter= biden_df["age"]>=0
print(biden_newfilter)
biden_df= biden_df[biden_newfilter]
print(biden_df)
```

And now, about the scatter ?

```
biden_df.plot(kind="scatter", x="age", y="biden", alpha=0.3)
```

Perfect ! What can we analyse then, about this relationship ? We can see there is a kind of **polarization** of opinions. Indeed, many people answered 0 or 100 (or close to 100, around 70) when they've been asked "How close do you feel your opinions are with the Biden's candidacy ?". The majority of dots are around 0 or above 70, but a fewer of them, as we can see, is between 2 and 60: Respondents were really "hot" or "cold" for Biden, but **not** that much **nuanced**. 

### Now that we analysed this, we can make the same analysis by comparing **Trump's data**. 

```
trump_variables= ["V202144","V201507x"]
trump_df= df[trump_variables]
trump_df.columns= ["trump", "age"]
trump_df.plot(kind= "scatter", x="age", y="trump", alpha= 0.1)
```

```
trump_filter= trump_df["trump"]>=0
print(trump_filter)
trump_df= trump_df.loc[trump_filter]
print(trump_df)
```

```
trump_df.plot(kind= "scatter", x="age", y="trump", alpha= 0.3)
```

```
trump_newfilter= trump_df["age"]>=0
print(trump_newfilter)
trump_df= trump_df.loc[trump_newfilter]
print(trump_df)
```

```

trump_df.plot(kind= "scatter", x="age", y="trump", alpha= 0.3)
```

Now that we did the same job for Trump's part --> it seems, at first glance, that they are the same scatter points actually. Maybe we can make something to distinguish them, based on the **traditional colors** of their respective political parties. 

```
trump_df.plot(kind="scatter", x="age", y="trump", alpha= 0.3, color='red')
```

```
trump_df.plot(kind="scatter", x="age", y="trump", alpha= 0.3, color='red')
biden_df.plot(kind= "scatter", x="age", y= "biden", alpha= 0.3)
```

Now that we've got both of them, it seems quite difficult to see differences between these variables relationships. At least, we can make 2 different conclusions. First of all: there is a large polarization between the lowest and the highest parts of both scatters. Secondly, we can at least see that respondents were **more nuanced about Trump** than they were with Biden, as we can see more "intermediate" results than for Biden's feeling thermometer. To conclude with, is there a difference between ages ? 

Let's try with another method, with a **"best fit line"**. This method can allow us to check, with an approximation, if there is a correlation or not between these variables. Visually, the best fit line will be made with as much points above this line than below this line. 

To use this method, we have to import the **"seaborn library"**

```
import seaborn as sns
## Now, let's use lmplot, which means linear model plot, with both of our scatterplots)
sns.lmplot(x="age", y="biden", data=biden_df)
```

```
sns.lmplot(x="age", y="trump", data=trump_df)
```

We can draw a better conclusion now, using the best fit line. Biden's thermometer is not really evolving between the different respondent's ages. But with Trump's case, we can see something really interesting. Biden's thermometer seems to be almost the same with the different respondents' ages, but Trump's thermometer median value is lower when the respondents are younger !
