## The importance of recoding 

Some variables needs to be recoded to seems clearer and more explicit during the data visualization. 

First of all, as usually you have to import pandas and dowload the data you want to analyse.

```python
import pandas as pd

my_link = "https://raw.githubusercontent.com/datamisc/ts-2020/main/data.csv"

df = pd.read_csv(my_link, compression='gzip')

df
```

For this live demo i picked two variables which are the sex of respondants and the importance of religion in their life.  
To simplify the lecture of these variables we can rename the columns before recoding the variables in the rows. 

```python
mes_vars = ["V201600", "V201433"]
df = df[mes_vars]
df

df.columns=["sex", "religion"]
df
```

```python
df.value_counts("religion")
#or
df["religion"].value_counts()
```

```python
df["sex"].value_counts()
```

As we can see the responses take the form of number which is not really explicit. It needs some simplification. 

Let's start by removing the results that are inferior at 0 and are inoperative, they are shown with the value_counts function. To do that task we are going to create and use a mask.

```python
mask=df["religion"] > 0
df = df[mask]
df["religion"].value_counts()
```

```python
mask=df["sex"] > 0
df = df[mask]
df["sex"].value_counts()
```

##The readability of results

Here we can see that the results are not really understable if you don't know which number is related to which response.  
In some situation it can also be problematic because the computer is going believe that a variable is superior at another because of it's number. 

Therefore in that case we have to assign and rename the variable. 

###Various ways possible 

We've seen with Clara the loc function that permits to rename variables, I'm going to show you again what it is with my variable of respondant's gender.

In a second time we are going to talk about the "apply method" which can be used with the same purpose.

###The loc function

The first step of this function is to create a filter to rename the results based upon respondant's gender.



```python
male_filter = df["sex"] == 1
df[male_filter]
```

Then you can apply it, you will have to run the 2 next cases and run it again.   

```python
df.loc[male_filter,"sex"]
```

```python
df.loc[male_filter,"sex"]="male"
df["sex"].value_counts()
```

Now you have to do the same thing for the others results that will be the female filter and variable.

```python
female_filter = df["sex"] == 2
df[female_filter] 
```

```python
df.loc[female_filter,"sex"]
```

```python
df.loc[female_filter,"sex"]="female"
df["sex"].value_counts()
```

Now that we have cleaned up variables, we can do percentages and graphs which are clearer. 

```python
df["sex"].value_counts(normalize=True) * 100
```

When two variables are put into perspective and only one is clearly identifiable the problem is to understand what it's supposed to mean like this tab. 

```python
pd.crosstab(df["sex"], df["religion"], normalize=True) * 100
```

Now we are going to recode the variable of religion but in a different way than just before.

By the apply method we are going to create a function that is going to rename automatically a variable.

To use this method you will have to run, between each category, the "df["religion"] = df["religion"].apply(religion)" case.

```python
def super_religion(series):
  if series == 1 :
    return 'Extremly important'

  elif series == 2 :
    return 'Very important'

  elif series == 3 :
    return 'Moderatly important'

  elif series == 4 :
    return 'A little important'

  elif series == 5 :
    return 'Not important at all'

  else:
    return series
```
